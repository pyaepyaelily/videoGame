```python
dupi_data = steam_data[steam_data.duplicated("name", keep = False)]
dupi_data.sort_values(by="appid").head(n=10)
```

```python
print("Steam games with duplicate names:", len(dupi_data))
dupis = dupi_data["name"].unique()
print("Games with duplicated Names:", len(dupis))
print()

print("Count\t App IDs \t \t Name of Game\n")
for dupi in dupis:
    dupi_list = list(dupi_data[dupi_data["name"] == dupi]['appid'])
    print(len(dupi_list), "\t", dupi_list, "\t", dupi)
```
